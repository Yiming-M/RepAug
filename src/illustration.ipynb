# Reproducible Random Augmentation

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

import repaug

seed = 42
```

```python
img_path = os.path.join(os.getcwd(), "test_img.jpg")
with Image.open(img_path) as img_:
    img = img_.convert("RGB").copy()

plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

## Random Color Jittering

```python
RandomColorJitter = repaug.RandomColorJitter(seed=seed)
img_aug = RandomColorJitter(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Color Jittering")
plt.show()
```

## Random Crop

```python
RandomCrop = repaug.RandomCrop(seed=seed)
img_aug = RandomCrop(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Cropping")
plt.show()
```

```python
RandomResizedCrop = repaug.RandomResizedCrop(seed=seed)
img_aug = RandomResizedCrop(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Resized Cropping")
plt.show()
```

## Flip

```python
RandomHorizontalFlip = repaug.RandomHorizontalFlip(p=0.8, seed=seed)
img_aug = RandomHorizontalFlip(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Horizontal Flipping")
plt.show()
```

```python
RandomVerticalFlip = repaug.RandomVerticalFlip(p=0.8, seed=seed)
img_aug = RandomVerticalFlip(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Vertical Flipping")
plt.show()
```

## Random Gaussian Blur

```python
RandomGaussianBlur = repaug.RandomGaussianBlur(kernel_size=[5, 7, 9], seed=seed)
img_aug = RandomGaussianBlur(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Gaussian Blurring")
plt.show()
```

## Pepper and Salt

```python
Pepper = repaug.Pepper(num_peppers=100, seed=seed)
img_aug = Pepper(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Pepper Noise")
plt.show()
```

```python
Salt = repaug.Salt(num_salts=100, seed=seed)
img_aug = Salt(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Salt Noise")
plt.show()
```

## Random Rotate

```python
RandomRotation = repaug.RandomRotation(degrees=30, seed=seed)
img_aug = RandomRotation(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Rotating")
plt.show()
```

```python
Random90Rotation = repaug.Random90Rotation(seed=seed + 53)
img_aug = Random90Rotation(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Rotating")
plt.show()
```

```python
Random180Rotation = repaug.Random180Rotation(seed=seed)
img_aug = Random180Rotation(img)

plt.imshow(img_aug)
plt.axis("off")
plt.title("Random Rotating")
plt.show()
```
